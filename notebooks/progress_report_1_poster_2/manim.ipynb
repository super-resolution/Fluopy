```python
%%manim -qh Example
config.frame_width = 14
config.frame_height = 8
config.disable_caching = True

class Example(Scene):
    def construct(self):        
        vertices = ['S0', 'S1', 'T1', 'B', 'OF', 'T2', 'Fl']
        edges = [('S1', 'S0'), ('S1', 'T1'), ('T1', 'B'), ('T1', 'S0'), ('S0', 'S1'), 
                 ('S1', 'OF'), ('OF', 'S0'), ('T1', 'T2'), ('T2', 'S1'), ('S1', 'Fl'),
                 ('T1', 'OF')]
        g = DiGraph(vertices, edges, labels=True, layout={'S0': [-5, -3, 0], 'S1': [-5, 2, 0], 'T1': [0, 1, 0], 
        'B': [-1, -3, 0], 'OF': [-3, -3, 0], 'T2': [0, 3, 0], 'Fl': [-6, 3, 0]})
        
        vertices2 = ['S0', 'S1', 'T1', 'OF']
        edges2 = [('S0', 'S1'), ('S1', 'T1'),  
                 ('S1', 'OF'), ('T1', 'OF')]
        g2 = DiGraph(vertices2, edges2, labels=True, layout={'S0': [-5, -3, 0], 'S1': [-5, 2, 0], 'T1': [0, 1, 0], 
        'OF': [-3, -3, 0]}) 

        periodicity = Text('1. Aperiodic?', color="#DB7921", font_size=29).move_to([3, 3, 0])
        absorb = Text('2. Absorbing states?', color="#DB7921", font_size=29).move_to([3, 2, 0])
        most_occ = Text('3. Simple cycles?', color="#DB7921", font_size=29).move_to([3, 1, 0])
        independent = Text('4. Independent?', color="#DB7921", font_size=29).move_to([3, 0, 0])
        topo_sort = Text('Topological sort', color="#DB7921", font_size=29).move_to([3, 2, 0])
        dag = Text('DAG: directed acyclic graph', color="#DB7921", font_size=29).move_to([3, 3, 0])

        self.camera.background_color = "#081017"
        self.add(g)
        self.play(FadeIn(periodicity))
        self.wait(2)
        self.play(FadeIn(absorb))
        self.wait(2)
        self.play(g['B'][0].animate.set_color('#BF3A2B'), g['B'][1].animate.set_color(BLACK),
                  g.edges[('T1', 'B')].animate.set_color('#BF3A2B'))
        self.play(FadeOut(g['B']), FadeOut(g.edges[('T1', 'B')]))
        self.play(FadeIn(most_occ))
        self.wait(1)
        self.play(Indicate(g['S0'], color='#E4D17E'), Indicate(g['S1'], color='#E4D17E'),
                  Indicate(g.edges[('S0', 'S1')], color='#E4D17E'), Indicate(g.edges[('S1', 'S0')], color='#E4D17E'),
                  run_time=2)
        self.play(Indicate(g['S0'], color='#E4D17E'), Indicate(g['S1'], color='#E4D17E'), Indicate(g['T1'], color='#E4D17E'),
                  Indicate(g.edges[('S0', 'S1')], color='#E4D17E'), Indicate(g.edges[('S1', 'T1')], color='#E4D17E'),
                  Indicate(g.edges[('T1', 'S0')], color='#E4D17E'),
                  run_time=2)
        self.play(Indicate(g['S0'], color='#E4D17E'), Indicate(g['S1'], color='#E4D17E'), Indicate(g['OF'], color='#E4D17E'),
                  Indicate(g.edges[('S0', 'S1')], color='#E4D17E'), Indicate(g.edges[('S1', 'OF')], color='#E4D17E'),
                  Indicate(g.edges[('OF', 'S0')], color='#E4D17E'),
                  run_time=2)
        self.play(Indicate(g['S0'], color='#E4D17E'), Indicate(g['S1'], color='#E4D17E'), Indicate(g['OF'], color='#E4D17E'),
                  Indicate(g.edges[('S0', 'S1')], color='#E4D17E'), Indicate(g.edges[('T1', 'OF')], color='#E4D17E'),
                  Indicate(g.edges[('OF', 'S0')], color='#E4D17E'), Indicate(g['T1'], color='#E4D17E'),
                  Indicate(g.edges[('S1', 'T1')], color='#E4D17E'),
                  run_time=2)
        self.play(Indicate(g['S1'], color='#E4D17E'), Indicate(g['T1'], color='#E4D17E'), Indicate(g['T2'], color='#E4D17E'),
                  Indicate(g.edges[('S1', 'T1')], color='#E4D17E'), Indicate(g.edges[('T1', 'T2')], color='#E4D17E'),
                  Indicate(g.edges[('T2', 'S1')], color='#E4D17E'),
                  run_time=2)
        self.play(FadeOut(g['T2']), FadeOut(g.edges[('T1', 'T2')]), FadeOut(g.edges[('T2', 'S1')]))
        self.play(FadeIn(independent))
        self.play(g['Fl'][0].animate.set_color('#BF3A2B'), g['Fl'][1].animate.set_color(BLACK),
                  g.edges[('S1', 'Fl')].animate.set_color('#BF3A2B'))
        self.play(FadeOut(g['Fl']), FadeOut(g.edges[('S1', 'Fl')]))
        self.wait(1)
        self.remove(independent, absorb, periodicity, most_occ)
        self.wait(1)
        self.play(FadeIn(dag))
        self.wait(2)
        self.remove(g['S0'], g['S1'], g['T1'], g['OF'], g.edges[('S0', 'S1')],
                    g.edges[('S1', 'T1')], g.edges[('S1', 'OF')], g.edges[('T1', 'OF')],
                    g.edges[('S1', 'S0')], g.edges[('OF', 'S0')], g.edges['T1', 'S0'])
        self.add(g2)
        self.wait(2)
        self.play(FadeIn(topo_sort))
        self.play(g2['S0'].animate.move_to([-5, -2, 0]),
                  g2['S1'].animate.move_to([-3, -2, 0]),
                  g2['T1'].animate.move_to([-1, -1, 0]),
                  g2['OF'].animate.move_to([1, -2, 0]), run_time=2)
        self.play(FadeIn(Text('1', color="#DB7921", font_size=29).next_to(g2['S0'], UP)),
                  FadeIn(Text('2', color="#DB7921", font_size=29).next_to(g2['S1'], UP)),
                  FadeIn(Text('3', color="#DB7921", font_size=29).next_to(g2['T1'], UP)),
                  FadeIn(Text('4', color="#DB7921", font_size=29).next_to(g2['OF'], UP)))


```
