# <a id='toc1_'></a>[Photoswitching fingerprints: 4 fluorophores - RISC (0.1 % of STA)](#toc0_)

```python
import numpy as np

import fluopy.distributions as dist
import fluopy.figure as fi
import fluopy.fluorophores as fl
import fluopy.transitions as tr
import fluopy.routines as rt

%load_ext autoreload
%autoreload 2

import warnings


def custom_warning_format(msg, category, filename, lineno, line=None):
    if line is None:
        import linecache

        line = linecache.getline(filename, lineno)
    if "Floating point precision" in str(msg):
        return ""
    return f"WARNING for line: {line} {msg} \n"


warnings.formatwarning = custom_warning_format

saving_at = r"D:\python_output\Chapter_I\1_8_multi_f_PFA_RISC\01"
```

## <a id='toc1_1_'></a>[Data generation](#toc0_)

```python
PARAMS_ADJ = {
    k: v for k, v in rt.PARAMS_DSTORM.items() if k != "energy_transfer_parameters"
}


def prepare_transition_set(distance):
    fluorophores = fl.construct_fluorophores(
        name="cy5_dna", count=4, distance=distance, shape="square"
    )
    fluorophore_system = fl.FluorophoreSystem(fluorophores=fluorophores)
    transitions = fluorophore_system.load_transitions(
        bleaching=True,
        summarize=True,
        energy_transfer=True,
        **PARAMS_ADJ,
        energy_transfer_parameters={
            "overwrite": {"t1": [1, 0.001]},
            "exclude": ["off", "s0"],
        }
    )
    transition_set = tr.TransitionSet(transitions, fluorophore_system)
    transition_set.finalize()

    return transition_set
```

### <a id='toc1_1_1_'></a>[3 nm](#toc0_)

```python
rng = np.random.default_rng(1)
transition_set = prepare_transition_set(distance=3)
_, _, _ = rt.fingerprint_analysis(
    transition_set,
    batches=1,
    batch_size=40,
    filepath=saving_at + r"\3nm",
    filename="RISC_3nm",
    seed=rng,
)
```
