# <a id='toc1_'></a>[Method example](#toc0_)

```python
import matplotlib.pyplot as plt

saving_at = r"D:\python_output\Chapter_I\1_0_method_example"
```

```python
fig, ax = plt.subplots(figsize=(18, 14))
ax.axis("off")


def draw_fluorophore(x, y, ax):
    # photophysical states
    ax.text(s=r"$S_{0}$", x=0.45 + x, y=0.45 + y, fontsize=20, ha="center", va="center")
    ax.text(s=r"$S_{1}$", x=0.45 + x, y=0.55 + y, fontsize=20, ha="center", va="center")
    ax.text(s=r"$T_{1}$", x=0.55 + x, y=0.5 + y, fontsize=20, ha="center", va="center")

    prop = dict(
        arrowstyle="-|>, head_width=0.4, head_length=0.8",
        shrinkA=0,
        shrinkB=0,
        facecolor="black",
        edgecolor="black",
        lw=1,
    )
    # s0 to s1
    ax.annotate(
        "", xy=(0.445 + x, 0.525 + y), xytext=(0.445 + x, 0.475 + y), arrowprops=prop
    )
    ax.text(s="1", x=0.43 + x, y=0.5 + y, fontsize=17, ha="center", va="center")
    # s1 to s0
    ax.annotate(
        "", xy=(0.455 + x, 0.475 + y), xytext=(0.455 + x, 0.525 + y), arrowprops=prop
    )
    ax.text(s="2,3", x=0.47 + x, y=0.5 + y, fontsize=17, ha="center", va="center")
    # s1 to t1
    ax.annotate(
        "", xy=(0.53 + x, 0.5 + y), xytext=(0.47 + x, 0.55 + y), arrowprops=prop
    )
    ax.text(s="4", x=0.5 + x, y=0.54 + y, fontsize=17, ha="center", va="center")
    # t1 to s0
    ax.annotate(
        "", xy=(0.47 + x, 0.45 + y), xytext=(0.53 + x, 0.5 + y), arrowprops=prop
    )
    ax.text(s="5", x=0.5 + x, y=0.49 + y, fontsize=17, ha="center", va="center")
    # draw rectangle around the fluorophore
    ax.add_patch(
        plt.Rectangle((0.4 + x, 0.4 + y), 0.2, 0.2, fill=None, edgecolor="black", lw=1)
    )


####### 1 F #######
draw_fluorophore(-0.4, 0.1, ax)
ax.text(
    s=r"$P=\{S_{0}, S_{1}, T_{1}\}$",
    x=0.1,
    y=0.4,
    fontsize=20,
    ha="center",
    va="center",
)
ax.text(
    s=r"$J=\{1, 2, 3, 4, 5\}$", x=0.1, y=0.35, fontsize=20, ha="center", va="center"
)
ax.text(
    s=r"$S=\{S_{0}, S_{1}, T_{1}\}$",
    x=0.1,
    y=0.3,
    fontsize=20,
    ha="center",
    va="center",
)
ax.text(
    s=r"$U=\{1, 2, 3, 4, 5\}$", x=0.1, y=0.25, fontsize=20, ha="center", va="center"
)
####### 2 F #######
draw_fluorophore(0.1, 0.1, ax)
draw_fluorophore(0.31, 0.1, ax)
ax.text(
    s=r"$P=\{S_{0}, S_{1}, T_{1}\}$",
    x=0.7,
    y=0.4,
    fontsize=20,
    ha="center",
    va="center",
)
ax.text(
    s=r"$J=\{1, 2, 3, 4, 5\}$", x=0.7, y=0.35, fontsize=20, ha="center", va="center"
)
ax.text(
    s=r"$S=\{S_{0}|S_{0}, S_{0}|S_{1}, S_{0}|T_{1}, S_{1}|S_{0}, T_{1}|S_{0},$"
    r"$S_{1}|S_{1}, S_{1}|T_{1}, T_{1}|S_{1}, T_{1}|T_{1}\}$",
    x=0.7,
    y=0.3,
    fontsize=20,
    ha="center",
    va="center",
)
ax.text(
    s=r"$U=\{1_{S_{0}|S_{0} \to S_{0}|S_{1}}, 1_{S_{0}|S_{0} \to S_{1}|S_{0}}$"
    r"$1_{S_{0}|S_{1} \to S_{1}|S_{1}}, 1_{S_{1}|S_{0} \to S_{1}|S_{1}}\}$"
    r"$1_{S_{0}|T_{1} \to S_{1}|T_{1}}, 1_{T_{1}|S_{0} \to T_{1}|S_{1}},...\}$",
    x=0.7,
    y=0.25,
    fontsize=20,
    ha="center",
    va="center",
)
fig.savefig(
    saving_at + r"\method_example.png", dpi=300, bbox_inches="tight", pad_inches=0.1
)
```
