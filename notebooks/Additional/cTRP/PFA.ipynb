## <a id='toc1_1_'></a>[Reading data](#toc0_)

#### <a id='toc1_1_1_1_'></a>[2 fluorophores](#toc0_)

```python
fingerprints_2f_all = pd.Series(
    np.zeros(300001),
    np.round(np.linspace(0, 300, 300001), decimals=12),
    dtype=np.int32,
)
locs = pd.Series(
    np.zeros(300001),
    np.round(np.linspace(0, 300, 300001), decimals=12),
    dtype=np.int32,
)

for file in Path(reading_from).iterdir():
    if file.suffix == ".parquet" and "2f" in file.name:
        fingerprints_2f_all += pd.read_parquet(file).sum(axis=1)
        locs += (pd.read_parquet(file) > 0).sum(axis=1)
fingerprint = fingerprints_2f_all.cumsum() / fingerprints_2f_all.sum()
fingerprints_2f_all = fingerprint
locs = locs.cumsum() / locs.sum()
locs_2f_all = locs
```

#### <a id='toc1_1_1_2_'></a>[3 fluorophores](#toc0_)

```python
fingerprints_3f_all = pd.Series(
    np.zeros(300001),
    np.round(np.linspace(0, 300, 300001), decimals=12),
    dtype=np.int32,
)
locs = pd.Series(
    np.zeros(300001),
    np.round(np.linspace(0, 300, 300001), decimals=12),
    dtype=np.int32,
)
for file in Path(reading_from).iterdir():
    if file.suffix == ".parquet" and "3f" in file.name:
        fingerprints_3f_all += pd.read_parquet(file).sum(axis=1)
        locs += (pd.read_parquet(file) > 0).sum(axis=1)
fingerprint = fingerprints_3f_all.cumsum() / fingerprints_3f_all.sum()
fingerprints_3f_all = fingerprint
locs = locs.cumsum() / locs.sum()
locs_3f_all = locs
```
