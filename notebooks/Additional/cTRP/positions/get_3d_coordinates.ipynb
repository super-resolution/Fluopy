```python
import numpy as np
from pymol import cmd
```

```python
cmd.load(
    r"C:\Users\vie43sq\data_vincent\python_output\Additional\cTRP\positions\pse_files\cTRP-2.pse"
)
lysine1 = cmd.get_model("select1")
lysine2 = cmd.get_model("select2")
distances = np.zeros((len(lysine1.get_coord_list()), len(lysine2.get_coord_list())))
for i, pos1 in enumerate(lysine1.get_coord_list()):
    for j, pos2 in enumerate(lysine2.get_coord_list()):
        dist = np.linalg.norm(np.array(pos1) - np.array(pos2))
        distances[i, j] = dist
pos1 = lysine2.get_coord_list()[3]
pos2 = lysine1.get_coord_list()[3]
positions_cTRP_2 = np.array([pos1, pos2])
```

```python
cmd.load(r"C:\Users\vie43sq\Downloads\pdb-files-cTRP\pdb-files-cTRP\cTRP-3.pse")
lysine1 = cmd.get_model("select1")
lysine2 = cmd.get_model("select2")
lysine3 = cmd.get_model("select3")
pos1 = lysine2.get_coord_list()[3]
pos2 = lysine1.get_coord_list()[3]
pos3 = lysine3.get_coord_list()[3]
positions_cTRP_3 = np.array([pos1, pos2, pos3])
```

```python
cmd.load(r"C:\Users\vie43sq\Downloads\pdb-files-cTRP\pdb-files-cTRP\cTRP-4.pse")
lysine1 = cmd.get_model("select1")
lysine2 = cmd.get_model("select2")
lysine3 = cmd.get_model("select3")
lysine4 = cmd.get_model("select4")
pos1 = lysine2.get_coord_list()[3]
pos2 = lysine1.get_coord_list()[3]
pos3 = lysine3.get_coord_list()[3]
pos4 = lysine4.get_coord_list()[3]
positions_cTRP_4 = np.array([pos1, pos2, pos3, pos4])
```
