## <a id='toc1_1_'></a>[Note](#toc0_)
### <a id='toc1_1_1_'></a>[Assumptions](#toc0_)
The following regards all k-fluorophore fitting procedures done using Photoswitching_fingerprint_model. 
1. It assumes that a fluorophore k time of bleaching (starting from time of bleaching of fluorophore k-1) is fully described by a bias rate, a non-bias rate and a mixture factor p.
   - the bias and non-bias is explained in more detail below
   - Assumption that physical factors determining the parameters are constant
2. It assumes that sufficient runs (simulation) or clusters (experiment) make up the data; i.e., the model describes the data on ensemble-level, not instance-level.\
   2.1 in a single run, *multiple* photons are emitted during short periods of time separated by long periods of time *without any* photons. Such correlations in time are getting less significant with increasing number of runs.\
   2.2 in a single run, a fluorophore either bleaches during bias or during non-bias, whereas the model assumes a mixture of those. For example, a fluorophore may bleach during bias, such that its non-bias component should no longer be considered (whether if talking about bleaching times nor photon arrival times).\
   2.3 The global bleaching times represent photon arrival times only on the ensemble level. For example, if a fluorophore bleaches during non-bias, it still produces photons during bias. This is only important when using the terms global bleaching times and photon arrival times interchangeably.\
    **Special case**: Under the assumption that energy transfers do not (significantly) change the number of photons emitted per ON state², one could consider samples to represent ON states (not ON periods!). It can then be multiplied with a number drawn from a distribution representing the number of photons per ON state. This would fully cover 2.1. The number of ON states should be high enough such that it is indeed a mixture of runs. A few runs (e.g., 10) may suffice. This would cover 2.2.\
    ²This is the case for most energy transfers: it competes with fluorescence, however it also competes with transitions to the OFF state. There are exceptions, e.g., ET-induced RISC increases the probability of fluorescence per excitation (S0-S1-T1-T2-S1 instead of S0-S1-T1)
### <a id='toc1_1_2_'></a>[Bias and non-bias global bleaching rates in a 1 fluorophore system](#toc0_)
The 2-exponential components approximation assumes two things: that there are only 2 components, and that they are exponentially distributed.\
\
The **bias** of a *single fluorophore* system concerns the first ON state. In some cases, the fluorophore might bleach in the very first acquired triplet state, whereas in other cases, some cis lifetimes and other things are happening before that. Hence, the exponential distribution approximation of the bias does not work properly here. The exponential approximation of waiting times until photobleaching within the first ON state hence need photobleaching to have a *low probability*. Then, the times until photobleaching become heavy-tailed and approach an exponential distribution. Here is why: Each time the system can 'decide' to photobleach or not, it increases the number of 'trials' - Bernoulli trials. Hence, the number of trials $K$ until success is geometrically distributed with sucess probability $p$. The time between trials $T$ is distributed with unknown distribution D. As $p$ gets smaller, $K$ increases and the product $T_{tot} = K * T$ becomes dominated by $K$, such that the distribution becomes exponentially (continuous case) distributed with $\lambda = p/µ_T$.\
\
The **non-biased** part concerns the time after the first ON state (note that it is not strictly cut at the end of the ON state, because this would create a bias that there is an OFF state in the beginning). This part is approximately exponentially distributed, because the lifetimes of the OFF state, which are the main component of time progression, are exponentially distributed. Of course, this assumes only one *long-lived* photophysical OFF state to be available, or if there are multiple, that their lifetimes are on the same time scale. 
### <a id='toc1_1_3_'></a>[Photon arrival times, $\Delta$ bleaching times, ON state times](#toc0_)
#### Photon arrival times
The experimental and simulated photoswitching fingerprints result from an overlay of several measurements (experimental clusters or simulated runs) of photon arrival times. The photon arrival times can either be depicted as eCDF or ePDF. 
#### $\Delta$ bleaching times
The times when fluorophores photobleach is related to the time dependent probability of photon emission; e.g., if there is a higher chance of photon arrival in the beginning of a measurement, there will also be a higher chance of photobleaching since the fluorophore(s) are more likely to be excited. The global bleaching rate is dependent on the ON state probability and on the probability to photobleach per ON state. At first this may seem unintuitive, because the number of photons in a time interval are much more than the number of bleaching events in the same time interval. However, this is more an issue of the sample size - bleaching can happen only once per fluorophore, whereas the number of photons is much higher (i.e., it is like drawing from the same distribution but for bleaching events the sample size is 4 and for photon arrival times the sample size is 10000).
when photobleaching rate goes up, it means that fluorophores bleach earlier. If they bleach earlier, the times of photon arrival get lowered, because later photon arrivals are less likely. Hence, it goes hand in hand? 
Does trhe bleaching rate dictate the photon arrival rate?


The correlation between $r_p$ and $r_B$ goes both ways, i.e., if the bleaching rate goes up due to energy transfer producing higher excited states, the subsequent photon arrival rate goes down, leading to the 

This correlation goes both ways, for example, energy transfer can lead to shorter OFF states which increases the $r_p$ and $r_B$, however energy transfers could also lead to higher excited states which would increase $r_B$ even more.



The (delta) bleaching times histogram can be directly fitted with a PDF. If some assumptions hold, a fluorophore bleaches with 2 rates. One for the bias due to simulation and experiment starting in an all-ON state (S0). The other for the unbiased case when the probability of finding an ON state at a given time is constant. These rates are termed *global bleaching rate*. It is the inverse of the lifetime of a fluorophore starting from the previous bleaching event.

However, depending on the probability, the fluorophores do not actually bleach within the given time - therefore, the MLE procedure can also cover the number of runs without events. 

Still, in real experiments, the bleaching times in multi-fluorophore systems are not extractable - also, fitting the PDF directly may result in a noisy fit. However, the fingerprints of a *single fluorophore* directly resemble the CDF of bleaching times. These plots are not as noisy and hence may result in a better fit. 

The 2-rate-approximation results in a 2-component exponential mixture distribution.

```python
# this function mimics that ON can rescue OFF. but it has a constant probability of
# doing so in reality, the probability of being rescued can vary, for example if the ON
# state rescues an OFF state, one of the fluorophores goes back to OFF, to rescue again
# the remaining fluorophore doesnt have as much time left on average


def samples_OFF_lifetimes(n, size, p, seed):
    rng = np.random.default_rng(seed)
    samples = []
    sorted_samples = rng.exponential(1, n)
    breaker = False
    for i in range(size):
        if breaker:
            break
        sorted_samples = np.sort(sorted_samples)
        shortest = sorted_samples[0]
        sorted_samples = np.delete(sorted_samples, 0)
        true_or_false = rng.choice([True, False], (n - 1,), p=[p, 1 - p])
        while True in true_or_false:
            sorted_samples[true_or_false] = rng.exponential(1, np.sum(true_or_false))
            samples += [shortest] * np.sum(true_or_false)
            i += 1
            if i >= size:
                breaker = True
                break
            true_or_false = rng.choice([True, False], (n - 1,), p=[p, 1 - p])
        else:
            samples.append(shortest)
            sorted_samples = np.append(sorted_samples, rng.exponential(1, 1))
    return np.array(samples).flatten()
```

```python
for i in range(1, 5):
    samples = samples_OFF_lifetimes(n=i, size=10000, p=0.5, seed=42)
    result = kstest(samples, "expon", args=(0, np.mean(samples)))
    print(result)
```

```python
identifiers = ["1e1", "1e2", "1e3", "1e4"]
bleaching_times_all = []
fingerprints_all = []
fingerprints_pdf_all = []

for _ in identifiers:
    fingerprints_all.append(
        pd.Series(
            np.zeros(300001),
            np.round(np.linspace(0, 300, 300001), decimals=12),
            dtype=np.int32,
        )
    )
    fingerprints_pdf_all.append(
        pd.Series(
            np.zeros(300001),
            np.round(np.linspace(0, 300, 300001), decimals=12),
            dtype=np.float64,
        )
    )
folder_path = r"C:\Users\vie43sq\Desktop\Simulations\simulation_data\bl_rate"
for file in glob.glob(folder_path + "/*"):
    if file.endswith(".npy"):
        bleaching_times_all.append(np.load(file))
    elif file.endswith(".parquet"):
        for id in identifiers:
            if id in file:
                fingerprints_all[identifiers.index(id)] += pd.read_parquet(file).sum(
                    axis=1
                )
                break

for i, fingerprint in enumerate(fingerprints_all):
    fingerprints_pdf_all[i] = fingerprint
    fingerprint = fingerprint.cumsum() / fingerprint.sum()
    fingerprints_all[i] = fingerprint
```

## <a id='toc1_7_'></a>[Connection of (non-)bias rate with ON and OFF lifetimes](#toc0_)

For a *single fluorophore*, the **bias** lifetime component $\tau_{b}$ is dictated mainly by ON state (i.e., ON period) lifetime $\tau_{ON}$. If the chance of bleaching per ON period is low, $\tau_{b}$ will be a higher than $\tau_{ON}$ because the longer an ON state, the more often the fluorophore had the chance to photobleach. I.e., the biased bleaching events will be happening in long ON states in the beginning. If the chance of bleaching per ON period is high, $\tau_{b}$ will eventually be smaller than $\tau_{ON}$ together with a large p for this component, because even the short ON periods will encounter enough bleaching possibilities. \
The **non-bias** lifetime component $\tau_{nb}$ is dictated by the OFF state (i.e., OFF period) lifetime $\tau_{OFF}$. If the chance of bleaching per ON period is low $\tau_{nb}$ will be higher than $\tau_{OFF}$, because on average many OFF periods appear before photobleaching. If the chance of bleaching per ON period is high, $\tau_{nb}$ will be roughly equal to $\tau_{OFF}$, because there is a high chance of photobleaching within the first ON state that occurs in the **non-biased** time period (i.e., exactly one OFF state occurs before that, and $\tau_{OFF}>>\tau_{ON}$).
