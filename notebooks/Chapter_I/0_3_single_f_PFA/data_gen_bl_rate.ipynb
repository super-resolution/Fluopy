# <a id='toc1_'></a>[Photoswitching fingerprints: modulating bleaching rate](#toc0_)

```python
import numpy as np

import fluopy.fluorophores as fl
import fluopy.routines as rt
import fluopy.transitions as tr

%load_ext autoreload
%autoreload 2

saving_at = r"D:\python_output\Chapter_I\0_3_single_f_PFA\bl_rate"
```

## <a id='toc1_1_'></a>[Data generation](#toc0_)

```python
fluorophores = fl.construct_fluorophores(name="cy5_dna", count=1)
fluorophore_system = fl.FluorophoreSystem(fluorophores=fluorophores)
transitions = fluorophore_system.load_transitions(
    bleaching=True,
    **rt.PARAMS_DSTORM,
    summarize=True,
)
transition_set = tr.TransitionSet(transitions, fluorophore_system)
transition_set.finalize()
```

### <a id='toc1_1_1_'></a>[Bleaching rate: $10^{1}$](#toc0_)

```python
rng = np.random.default_rng(1)
transition_set = transition_set.adjust_rates({8: 1e1})
transition_set.finalize()
fingerprint_1e1, bl_times_1e1, _ = rt.fingerprint_analysis(
    transition_set,
    batches=1,
    batch_size=500,
    filepath=saving_at + r"\1e1",
    filename="bl_rate_1e1",
    seed=rng,
)
```
