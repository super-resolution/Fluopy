## Lambert-Beer-Law 

The Lambert-Beer-Law describes the absorption of a molecule depending on the intensity of the incident light, the concentration of the molecule and the path length the light has to travel through the homogenous sample. The transmitted light (and therefore the absorbed) light depends on these measures. The molecule-specific measure is the extinction coefficient $\epsilon$.

$A=\epsilon c d=\log(\frac{I_{0}}{I_{t}})$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eps = 2.5e5
d = 1 
c1 = 1e-5
c2 = 1e-7
c3 = 1e-9
c4 = 1e-11
concentrations = [c1, c2, c3, c4]
```

```python
I_0_list = np.arange(0, 20, 1)
I_transmitted_list = []
for c in concentrations:
    I_transmitted_list_c = []
    for I_0 in I_0_list:
        I_transmitted = I_0*10**(-eps*d*c)
        I_transmitted_list_c.append(I_transmitted)
    I_transmitted_list.append(I_transmitted_list_c)
```

The light transmitted through an absorbing sample can be described by an exponential decay. This is due to any absorbed light reduces the current intensity which decreases the decay rate. 

```python
[plt.plot(I_0_list, I_transmitted_list_, label=conc) for (I_transmitted_list_, conc) in zip(I_transmitted_list, concentrations)]
plt.legend(title='concentrations')
plt.ylabel(r"$I_t$")
plt.xlabel(r"$I_0$")
```

The incident intensity $I_{0}$ dictates the transmitted intensity linearly. The relationship depends on the extinction coefficient, path length and concentration of the sample. If the incident light has an intensity high enough to adress all molecules of the constant concentration, any additional incident light is fully transmitted. This is connected to the excitation rate. The higher the intensity of light, the higher the excitation rate. 

In this graph the incident intensity is constant but the concentration changes. If the concentration is high enough for a certain intensity, the light gets fully absorbed and the transmitted intensity approaches 0.

```python
concentrations = np.logspace(-10, -4, 10)
I_t_list = []
I_0 = 1e-6
eps = 2.5e5
d = 1 
for c in concentrations:
    I_t = I_0 * 10**(-eps*d*c)
    I_t_list.append(I_t)
    
plt.plot(concentrations, I_t_list)
plt.ylabel(r'$I_t$')
plt.xlabel(r'concentration')
```
