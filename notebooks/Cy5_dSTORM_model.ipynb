# Cy5 dSTORM model ($tS_0, tS_1, tT_1, cS_0, cS_1, cT_1, OFF, B$)
This model comprises the ground state $S_0$, the excited singlet state $S_1$, the excited triplet state $T_1$, all in the trans $t$ or cis $c$ configuration (Cy5 isomers) as well as the OFF state and the bleached state $B$.

### Define the state transition rates

```python
rate_dict = dict(k_tS0_tS1=[7e6, "excitation"],  
                 k_tS1_tS0=[1e9, "emission"],  
                 k_tS1_tT1=[1e6, "ISC"],   
                 k_tT1_tS0=[5e5, "ISC & VR"],  
                 k_tS1_B=[1e3, "bleaching"],
                 
                 k_cS0_cS1=[5e5, "excitation (c)"],
                 k_cS1_cS0=[1e9, "deexcitation (c)"],
                 k_tS1_cS0=[1e6, "ISO"],
                 k_cS1_tS0=[1e6, "BISO"],
                 
                 k_tT1_OFF=[1e7, "reduction"],
                 k_OFF_tS0=[1e4, "oxidation"])  
```

### Initialize system

```python
system = fs.Cy5dSTORMModel(number=2, distances=1, rates=rate_dict)
```

### Simulate the Markov process

```python
system.simulate(n_steps=1000, seed=99, base="py")
```

### Receive emission events

```python
system.emitters(unit="s", resample="0.005ms", photon_collection=1)
```

### Unique states and their occupation times

```python
system.process()
```

```python
system.fcs()
```
