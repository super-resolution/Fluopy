### 1 fluorophore, S0 S1 T1 R system

```python
rate_dict = dict(k_S0_S1=[1e15, "excitation"],
                 k_S1_S0=[1e9, "fluorescent emission"],
                 k_S1_T1=[1e6, "intersystem crossing"],
                 k_T1_S0=[5e5, "triplet deexcitation"],
                 k_T1_R=[1e4, "reduction"],
                 k_R_S0=[1e0, "oxidation"])
```

```python
system = fs.JablonskiModel(number=1,  distances=1, rates=rate_dict)
```

```python
system.simulate(n_steps=1000000, seed=99, base="py")
```
